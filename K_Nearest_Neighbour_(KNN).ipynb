# **Import Library**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# **Importing dataset**
Salah satu kumpulan data paling awal yang diketahui digunakan untuk mengevaluasi metode klasifikasi, dibuat oleh R. A. Fisher pada tahun 1936.

Nama Dataset : Iris

URL : https://archive.ics.uci.edu/dataset/53/iris

```python
# install library from source dataset
!pip install ucimlrepo
```

```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
iris = fetch_ucirepo(id=53)

# data (as pandas dataframes)
X = iris.data.features
y = iris.data.targets

#variable
iris.variables
```

# **The first five & the last five of dataset**

```python
# cek data column
dataset = iris.data.original
dataset
```

# **Preprocessing**

```python
# split into independent (x) and dependent (y) variable
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,4].values
```

# **Train test split**

```python
from sklearn.model_selection import train_test_split

#split data into x train, y train, x test and y test with 10% of data for testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30)
```

# **Feature Scaling**
Sebelum membuat prediksi, lebih baik dilakukan skala pada data untuk mendapatkan kecepatan yang lebih baik.
Kali ini kita menggunakan StandardScaler untuk menskala data dari -1 sampai 1.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x_train)

x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)

x_test
```

# **Training and Predictions**
Kita menggunakan algoritma KNN untuk melakukan prediksi dengan alat [scikit-learn](https://scikit-learn.org).

```python
from sklearn.neighbors import KNeighborsClassifier

# training data
classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(x_train, y_train)
```

```python
# prediction data
y_pred = classifier.predict(x_test)

# show predict data
y_pred
```

```python
# show test data
y_test
```

# **Evaluating the algorithm**

```python
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

# show result for evaluation algorithm with confusion matrix
confusionMatrix = confusion_matrix(y_test,y_pred)
classificationReport = classification_report(y_test,y_pred)
print(confusionMatrix)
print(classificationReport)
cmDisplay = ConfusionMatrixDisplay(confusion_matrix=confusionMatrix, display_labels=classifier.classes_)
cmDisplay.plot()
plt.show()
```
