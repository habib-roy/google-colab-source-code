### 2.1. Diberikan citra warna sebagai berikut dengan urutan warna R G B.

```python
# terdapat matrik citra 2bit dengan urutan warna R G B.
matrikCitra = np.array([
    [[3, 3, 3], [2, 1, 2], [2, 1, 3], [0, 0, 1]],
    [[2, 1, 0], [3, 1, 0], [1, 0, 3], [3, 1, 2]],
    [[3, 0, 1], [3, 3, 2], [1, 1, 1], [0, 1, 3]]
])
plt.imshow(matrikCitra)
plt.show()
```

### 2.2. Jika diinginkan untuk merubah citra tersebut menjadi citra grayscale dengan standar NTSC(0.299R + 0.587G + 0.114B), gambarkan matrik citra yang terbentuk!

```python
# cek matrix shape data asli citra
(row, col, channel) = matrikCitra.shape
print('matrix shape data asli citra (row, col, channel) : ', ( row, col, channel) )
```

image.png

```python
# deklarasi variable untuk imageGrayscale /pixel
imageGrayscale = np.zeros((row, col, 1)) # 1 channel karena untuk grayscale

# konversi matrik citra menjadi grayscale dengan standar NTSC(0.299R + 0.587G + 0.114B)
# https://id.wikipedia.org/wiki/NTSC
for r in range(row):
  for c in range(col):
    R = matrikCitra[r, c, 0] # Red pada index ke 0
    G = matrikCitra[r, c, 1] # Green pada index ke 1
    B = matrikCitra[r, c, 2] # Blue pada index ke 2

    grayValue = R*0.299 + G*0.587 + B*0.114 # NTSC
    imageGrayscale[r,c] = grayValue

print('matrix grayscale dengan standar NTSC', imageGrayscale)
```

```python
# menampilkan perbandingan gambar asli dan hasil konversi
plt.figure(figsize=(12, 4))

# gambar asli
plt.subplot(1, 3, 1)
plt.imshow(matrikCitra)
plt.title('Citra Asli (RGB)')

# gambar hasil konversi dengan colormap default (viridis)
# https://matplotlib.org/stable/users/explain/colors/colormaps.html
plt.subplot(1, 3, 2)
plt.imshow(imageGrayscale) # cmap=viridis
plt.title('Citra Hasil Konversi NTSC (Asli)')

# gambar hasil konversi dengan colormap grayscale
# https://matplotlib.org/3.5.3/api/_as_gen/matplotlib.pyplot.gray.html
plt.subplot(1, 3, 3)
plt.imshow(imageGrayscale, cmap='gray')
plt.title('Citra Hasil Konversi NTSC (Grayscale)')

plt.show()
```

### 3.1. Gambarkan histogram warna citra 1

```python
# deklarasi dataset
citra1 = np.array([
    [[3, 2, 0, 3]],
    [[1, 2, 2, 0]],
    [[0, 2, 0, 0]],
    [[2, 3, 2, 3]]
])

citra2 = np.array([
    [[2, 2, 0, 3]],
    [[2, 2, 2, 0]],
    [[2, 1, 0, 0]],
    [[2, 1, 2, 3]]
])

citra3 = np.array([
    [[1, 2, 0, 2]],
    [[1, 1, 2, 0]],
    [[0, 1, 1, 0]],
    [[2, 3, 1, 1]]
])
```

```python
# deklarasi fungsi untuk menghitung histogram
def hitungHistogram(x):
  xMin = np.min(x)
  xMax = np.max(x)
  countHistogram = []
  xData = []
  (row, col, channel) = x.shape
  for i in range(xMin, xMax+1):
    countHistogram.append(0)
    xData.append(i)
    for r in range(row):
      for c in range(col):
        for ch in range(channel):
          if(i == x[r,c,ch]): countHistogram[i] = countHistogram[i] + 1

  return xData, countHistogram
```

```python
import seaborn as sns

# menampilkan perbandingan gambar dengan histogram
plt.figure(figsize=(16, 4))

# gambar citra 1
plt.subplot(1, 4, 1)
plt.imshow(citra1)
plt.title('Citra 1')

# mendapatkan histogram citra 1
plt.subplot(1, 4, 2)
( xData1, histogramCitra1 ) = hitungHistogram(citra1)
print('Histogram Citra 1', histogramCitra1)
sns.barplot(x=xData1, y=histogramCitra1)
plt.title('Histogram Citra 1')

# mendapatkan histogram citra 2
plt.subplot(1, 4, 3)
( xData2, histogramCitra2 ) = hitungHistogram(citra2)
print('Histogram Citra 2', histogramCitra2)
sns.barplot(x=xData2, y=histogramCitra2)
plt.title('Histogram Citra 2')

# menampilkan hasil histogram
plt.subplot(1, 4, 4)
( xData3, histogramCitra3 ) = hitungHistogram(citra3)
print('Histogram Citra 3', histogramCitra3)
sns.barplot(x=xData3, y=histogramCitra3)
plt.title('Histogram Citra 3')

plt.show()
```

### 3.2. Dengan menggunakan jarak histogram (Eucledian) cari diantara citra2 dan citra 3 mana citra yang paling mirip dengan citra 1

```python
# deklarasi fungsi untuk menghitung nilai SSW
def hitungSSW(x):
  averageX = np.average(x) # hitung centroid
  jarakPerPixelDenganCenter = [] # deklarasi variabel untuk menampung array hasil hitung jarak
  (row, col, channel) = x.shape
  for r in range(row):
    for c in range(col):
      for ch in range(channel):
        ###
        # Menghitung jarak pixel dengan centroid
        # lalu dikuadratkan
        # lalu diakarkuadrat
        # dan dikumpulkan pada variable array hasil
        ###
        jarakPerPixelDenganCenter.append(np.sqrt(np.absolute(averageX - x[r,c,ch])**2))

  return (1/channel) * np.sum(jarakPerPixelDenganCenter)

# menghitung nilai SSW Citra 1
citra1SSW = hitungSSW(citra1)
print('SSW Citra 1 : ', citra1SSW)

# menghitung nilai SSW Citra 2
citra2SSW = hitungSSW(citra2)
print('SSW Citra 2 : ', citra2SSW)

# menghitung nilai SSW Citra 3
citra3SSW = hitungSSW(citra3)
print('SSW Citra 3 : ', hitungSSW(citra3))
```

```python
# deklarasi fungsi untuk menghitung SSB
def hitungSSB(x,y):
  xAverage = np.average(x)
  yAverage = np.average(y)
  return np.sqrt((np.absolute(xAverage - yAverage))**2)

# menghitung nilai SSB antara Citra 1 dengan Citra 2
citra12SSB = hitungSSB(citra1, citra2)
print('SSB Citra 1 - 2: ', citra12SSB)

# menghitung nilai SSB antara Citra 1 dengan Citra 3
citra13SSB = hitungSSB(citra1, citra3)
print('SSB Citra 1 - 3: ', citra13SSB)

# menghitung nilai SSB antara Citra 2 dengan Citra 3
citra23SSB = hitungSSB(citra2, citra3)
print('SSB Citra 2 - 3: ', citra23SSB)
```

```python
# hitung R Citra 1 - 2
R12 = (citra1SSW + citra2SSW) / citra12SSB
print('Jarak citra 1 dengan citra 2 : ', R12)

# hitung R Citra 1 - 3
R13 = (citra1SSW + citra3SSW) / citra13SSB
print('Jarak citra 1 dengan citra 3 : ', R13)

# hitung R Citra 2 - 3
R23 = (citra2SSW + citra3SSW) / citra23SSB
print('Jarak citra 2 dengan citra 3 : ', R23)
```

```python
# jarak terdekat dari citra 1 adalah
citraTerdekatDariCitra1 = "Citra 3" if (R12 > R13) else "Citra 2"

print('Citra Terdekat Dari Citra 1 adalah', citraTerdekatDariCitra1)
```

### 3.3. Berapa besar moment warna order 1, energy, contras dan homogenitas dari citra 1

```python
# momen warna order 1 adalah Mean
def colorMomentMean(x):
  meanAll = [] # all data rata-rata /channel
  (row, col, channel) = x.shape
  for ch in range(channel):
    chan = [] # data /channel
    for r in range(row):
      for c in range(col):
        chan.append(x[r,c,ch])
    meanPerChannel = np.sum(chan) / len(chan)
    meanAll.append(meanPerChannel) # data mean /channel, push to all data

  return np.sum(meanAll) / len(meanAll) # data mean all
```

6.1. Dengan menggunakan python tampilkan salah satu bentuk visualisasi dari hasil clustering data

```python
# mengambil dataset dari (kaggle.com) soal nomor 5.1.2 Hepatitis C Virus (HCV)
dataset.info()

# preview isi data
dataset
```

```python
# hapus semua kolom kecuali RNA Base & RNA 4
finalDataset = dataset[['RNA Base', 'RNA 4']]

# preview isi data
finalDataset.info()
finalDataset
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# pre-processing
dataTerskala = StandardScaler().fit_transform(finalDataset)

# data visualisasi awal
plt.scatter(dataTerskala[:, 0], dataTerskala[:, 1]);
```

```python
# mencari jumlah kluster paling optimal
sse = []
for k in range(1, 11):
  kmeans = KMeans(n_init=10, n_clusters=k)
  kmeans.fit(dataTerskala)
  sse.append(kmeans.inertia_)

print(sse)
# menampilkan Sum of Square Error (SSE)
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Jumlah kluster")
plt.ylabel("Sum of Square Error (SSE)")
plt.show()
```

```python
# menampilkan hasil clustering
hasilKmeans = KMeans(n_init=10, n_clusters=4)  # 4 kluster
hasilKmeans.fit(dataTerskala)
yKmeans = hasilKmeans.predict(dataTerskala)
plt.scatter(dataTerskala[:,0], dataTerskala[:,1], c=yKmeans, s=50, cmap='rainbow');
```
